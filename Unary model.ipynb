```python
#single_class_svm_poly(0.0001,1,'10_largest',train_cough,test_cough,test_snore)

def single_class_rfb(nu,gamma,variance_order,features,train_cough,test_cough,test_snore):
    test_outlier = test_snore
#    X = train_cough[features]
#     X_test = test_cough[features]
#     initial_features = train_cough.columns
#     features = list(initial_features)[3:43]
#     X = train_cough[features]
#     X_test = test_cough[features]

#     if variance_order =='10_largest':
#         new_features = X.std().nlargest(20).index.values
#     elif variance_order =='10_smallest':
#         new_features = X.std().nsmallest(20).index.values

#     X_train = X[new_features]
#     X_test = X_test[new_features]
#     X_outliers = test_outlier[new_features]

    X_train = train_cough[['delta_32', 'delta_38', 'delta_33' ,'delta_39' ,'delta_34' ,'delta_35'
 ,'delta_37', 'delta_36', 'delta_27', 'delta_31', 'delta_23', 'delta_25',
 'delta_24', 'delta_30', 'delta_18', 'delta_22', 'delta_28', 'delta_26',
 'delta_29' ,'delta_19']]

    X_test = test_cough[['delta_32' 'delta_38' 'delta_33' 'delta_39' 'delta_34' 'delta_35'
 'delta_37' 'delta_36' 'delta_27' 'delta_31' 'delta_23' 'delta_25'
 'delta_24' 'delta_30' 'delta_18' 'delta_22' 'delta_28' 'delta_26'
 'delta_29' 'delta_19']]
    X_outliers = test_snore[['delta_32' 'delta_38' 'delta_33' 'delta_39' 'delta_34' 'delta_35'
 'delta_37' 'delta_36' 'delta_27' 'delta_31' 'delta_23' 'delta_25'
 'delta_24' 'delta_30' 'delta_18' 'delta_22' 'delta_28' 'delta_26'
 'delta_29' 'delta_19']]

    
    
    clf = svm.OneClassSVM(nu=nu, kernel="rbf",gamma=gamma)
    clf.fit(X_train)

    y_pred_train = clf.predict(X_train)
    y_pred_test = clf.predict(X_test)
    y_pred_outliers = clf.predict(X_outliers)


    n_error_train = y_pred_train[y_pred_train == -1].size
    n_error_test = y_pred_test[y_pred_test == -1].size
    n_error_outliers = y_pred_outliers[y_pred_outliers == 1].size



    TN = len(X_outliers)-n_error_outliers 
    FN = n_error_test 
    TP = len(X_test)-n_error_test
    FP = n_error_outliers
    
    test_total_count = len(X_test)+len(X_outliers)
    error_rate_train = n_error_train/len(X_train)
    error_rate_test = n_error_test/len(X_test)
    error_rate_outlier = n_error_outliers/len(X_outliers)
    precision = TP/(TP+FP+0.0000001)
    recall = TP/(TP+FN+0.0000001)
    F1 = (2*precision*recall)/(precision+recall+0.0000001)
    ACC_True = TP/(TP+FN+0.0000001)
    ACC_False = TN/(FP+TN+0.0000001)
    acc_rate = (TP+TN)/(TP+TN+FN+FP)
    FPR = FP/(FP+TN)
    FNR = FN/(FN+TP)
    temp = {'ACC_True':ACC_True,'ACC_False':ACC_False ,'TP':TP,'FP':FP,'TN':TN,'FN':FN
            ,'precision':precision,'recall':recall,'F1':F1
            ,'gamma':gamma
            ,'nu':nu,'total_count':test_total_count
       ,'acc_rate':acc_rate,'FPR':FPR,'FNR':FNR}

    return TP,FP,TN,FN,error_rate_test,error_rate_outlier,error_rate_train,nu,test_total_count,gamma
```

```python
def single_class_svm_poly(nu,degree,variance_order,features,train_cough,test_cough,test_snore):

    test_outlier = test_snore
    X = train_cough[features]
    X_test = test_cough[features]
    if variance_order =='10_largest':
        new_features = X.std().nlargest(10).index.values
    elif variance_order =='10_smallest':
        new_features = X.std().nsmallest(10).index.values

    X_train = X[new_features]
    X_test = X_test[new_features]
    X_outliers = test_outlier[new_features]
    #fit model
    clf = svm.OneClassSVM(nu=nu, kernel="poly", degree=degree).fit(X_train)

    y_pred_train = clf.predict(X_train)
    y_pred_test = clf.predict(X_test)
    y_pred_outliers = clf.predict(X_outliers)
    
    n_error_train = y_pred_train[y_pred_train == -1].size
    n_error_test = y_pred_test[y_pred_test == -1].size
    n_error_outliers = y_pred_outliers[y_pred_outliers == 1].size
    
    TN = len(X_outliers)-n_error_outliers 
    FN = n_error_test 
    TP = len(X_test)-n_error_test
    FP = n_error_outliers
    test_total_count = len(X_test)+len(X_outliers)
    error_rate_train = n_error_train/len(X_train)
    error_rate_test = n_error_test/len(X_test)
    error_rate_outlier = n_error_outliers/len(X_outliers)
    
    return TP,FP,TN,FN,error_rate_test,error_rate_outlier,error_rate_train,nu,test_total_count,degree
```
